## 2.1 Isolation Forest

Isolation Forest isolates anomalies by randomly partitioning the feature space.
Anomalous points require **fewer splits** to isolate — they receive higher anomaly scores.

$$s(x, n) = 2^{-\frac{E(h(x))}{c(n)}}$$

Where $h(x)$ is the path length to isolate point $x$ and $c(n)$ is the expected path length for $n$ samples.
Score → 1.0 = anomalous | Score → 0.5 = normal.

````python
# ── Train Isolation Forest ────────────────────────────────────
print('🌲 Training Isolation Forest...')
print(f'   n_estimators  : {ISO_N_ESTIMATORS}')
print(f'   contamination : {ISO_CONTAMINATION}')
print(f'   Input shape   : {X_scaled.shape}\n')

iso_forest = IsolationForest(
    n_estimators=ISO_N_ESTIMATORS,
    contamination=ISO_CONTAMINATION,
    max_features=0.8,
    random_state=RANDOM_SEED,
    n_jobs=-1,
)
iso_forest.fit(X_scaled)

# score_samples returns negative anomaly scores
# More negative = more anomalous
# We negate and normalise to [0, 1]
iso_raw_scores = -iso_forest.score_samples(X_scaled)

mm_scaler_iso = MinMaxScaler()
iso_scores_norm = mm_scaler_iso.fit_transform(
    iso_raw_scores.reshape(-1, 1)
).flatten()

fm['iso_score'] = iso_scores_norm

print(f'✅ Isolation Forest trained.')
print(f'   Score range : [{iso_scores_norm.min():.4f}, '
      f'{iso_scores_norm.max():.4f}]')
print(f'   Score mean  : {iso_scores_norm.mean():.4f}')
print(f'   Flagged rows (score > {THRESHOLD_ELEVATED}): '
      f'{(iso_scores_norm > THRESHOLD_ELEVATED).sum():,}')
````

## 2.2 Autoencoder (PyTorch)

The Autoencoder learns a compressed representation of **normal** behavior.
When it tries to reconstruct an **anomalous** user's features, the reconstruction
error is high — that error becomes our anomaly signal.

```
Input(N) → Dense(32,ReLU) → Dense(16,ReLU) → Dense(8,ReLU)  ← bottleneck
         → Dense(16,ReLU) → Dense(32,ReLU) → Output(N)
```

**Key:** Trained **only** on users with low Isolation Forest scores (presumed normal).

````python
# ── Autoencoder architecture ──────────────────────────────────
class BehaviorAutoencoder(nn.Module):
    def __init__(self, input_dim: int, bottleneck: int = 8):
        super().__init__()
        self.encoder = nn.Sequential(
            nn.Linear(input_dim, 32),
            nn.ReLU(),
            nn.Dropout(0.1),
            nn.Linear(32, 16),
            nn.ReLU(),
            nn.Linear(16, bottleneck),
            nn.ReLU(),
        )
        self.decoder = nn.Sequential(
            nn.Linear(bottleneck, 16),
            nn.ReLU(),
            nn.Linear(16, 32),
            nn.ReLU(),
            nn.Dropout(0.1),
            nn.Linear(32, input_dim),
        )

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        return self.decoder(self.encoder(x))

    def reconstruction_error(
        self, x: torch.Tensor
    ) -> torch.Tensor:
        with torch.no_grad():
            recon = self.forward(x)
            return torch.mean((x - recon) ** 2, dim=1)


INPUT_DIM = X_scaled.shape[1]
autoencoder = BehaviorAutoencoder(
    input_dim=INPUT_DIM,
    bottleneck=AE_BOTTLENECK
).to(DEVICE)

print(f'✅ Autoencoder architecture:')
print(autoencoder)
total_params = sum(p.numel() for p in autoencoder.parameters())
print(f'\n   Total parameters : {total_params:,}')
print(f'   Input dimension  : {INPUT_DIM}')
print(f'   Bottleneck dim   : {AE_BOTTLENECK}')
````

````python
# ── Prepare training data (normal users only) ─────────────────
# Use bottom 80th percentile of ISO scores as 'normal' training set
normal_threshold = np.percentile(iso_scores_norm, 80)
normal_mask = iso_scores_norm <= normal_threshold

X_normal = X_scaled[normal_mask]
X_tensor  = torch.tensor(X_normal, dtype=torch.float32)
X_all_tensor = torch.tensor(X_scaled, dtype=torch.float32)

train_dataset = TensorDataset(X_tensor, X_tensor)
train_loader  = DataLoader(
    train_dataset,
    batch_size=AE_BATCH_SIZE,
    shuffle=True,
    drop_last=False,
)

print(f'✅ Autoencoder training set prepared.')
print(f'   Normal threshold (80th pct ISO): {normal_threshold:.4f}')
print(f'   Training samples : {X_normal.shape[0]:,} '
      f'({X_normal.shape[0]/len(X_scaled)*100:.1f}% of all rows)')
print(f'   Held-out rows    : {(~normal_mask).sum():,} '
      f'(high ISO score — not used for AE training)')
````

````python
# ── Train Autoencoder ─────────────────────────────────────────
criterion = nn.MSELoss()
optimizer = optim.Adam(autoencoder.parameters(), lr=AE_LR)
scheduler = optim.lr_scheduler.StepLR(
    optimizer, step_size=15, gamma=0.5
)

train_losses = []

print(f'🧠 Training Autoencoder for {AE_EPOCHS} epochs...\n')
autoencoder.train()

for epoch in range(1, AE_EPOCHS + 1):
    epoch_loss = 0.0
    for batch_x, batch_y in train_loader:
        batch_x = batch_x.to(DEVICE)
        optimizer.zero_grad()
        recon = autoencoder(batch_x)
        loss  = criterion(recon, batch_x)
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item() * len(batch_x)

    avg_loss = epoch_loss / len(X_normal)
    train_losses.append(avg_loss)
    scheduler.step()

    if epoch % 10 == 0 or epoch == 1:
        print(f'   Epoch {epoch:>3}/{AE_EPOCHS}  '
              f'Loss: {avg_loss:.6f}  '
              f'LR: {scheduler.get_last_lr()[0]:.6f}')

print(f'\n✅ Autoencoder training complete.')
print(f'   Final loss: {train_losses[-1]:.6f}')
````

````python
# ── Plot training loss curve ──────────────────────────────────
plt.figure(figsize=(10, 4))
plt.plot(range(1, AE_EPOCHS + 1), train_losses,
         color='#1E90FF', linewidth=2)
plt.fill_between(range(1, AE_EPOCHS + 1), train_losses,
                 alpha=0.2, color='#1E90FF')
plt.title('Autoencoder Training Loss (MSE)', fontsize=13,
          fontweight='bold')
plt.xlabel('Epoch')
plt.ylabel('MSE Loss')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('ae_training_loss.png', dpi=120, bbox_inches='tight')
plt.show()
print('✅ Training loss curve saved → ae_training_loss.png')
````

````python
# ── Compute reconstruction errors for all rows ────────────────
autoencoder.eval()

ae_errors = []
batch_size = 512
with torch.no_grad():
    for i in range(0, len(X_all_tensor), batch_size):
        batch = X_all_tensor[i:i+batch_size].to(DEVICE)
        err   = autoencoder.reconstruction_error(batch)
        ae_errors.append(err.cpu().numpy())

ae_errors_raw = np.concatenate(ae_errors)

mm_scaler_ae = MinMaxScaler()
ae_scores_norm = mm_scaler_ae.fit_transform(
    ae_errors_raw.reshape(-1, 1)
).flatten()

fm['ae_score'] = ae_scores_norm

print(f'✅ Autoencoder reconstruction errors computed.')
print(f'   Error range : [{ae_errors_raw.min():.6f}, '
      f'{ae_errors_raw.max():.6f}]')
print(f'   Norm range  : [{ae_scores_norm.min():.4f}, '
      f'{ae_scores_norm.max():.4f}]')
print(f'   Flagged rows (score > {THRESHOLD_ELEVATED}): '
      f'{(ae_scores_norm > THRESHOLD_ELEVATED).sum():,}')
````

## 2.3 LSTM Sequence Model (PyTorch)

The LSTM operates on a **30-day rolling window** of daily behavioral vectors per user.
An attention mechanism identifies **which days** within the window contributed most
to the anomaly score — directly explainable to SOC analysts.

```
Input: (batch, 30 days, N features)
  → LSTM(hidden=128, layers=2, dropout=0.2)
  → Attention over time steps
  → Dense(64, ReLU) → Dense(1, Sigmoid)
  → sequence_anomaly_score ∈ [0, 1]
```

**Training strategy:** We use reconstruction loss on normal sequences
(same principle as the Autoencoder — train only on low-risk users).

````python
# ── Build 30-day rolling windows per user ─────────────────────
print(f'🔧 Building {LSTM_WINDOW_DAYS}-day rolling windows per user...')

def build_sequences(
    df: pd.DataFrame,
    feature_cols: list,
    window: int
) -> tuple:
    """
    For each user, slide a window of `window` days over their
    daily feature vectors.

    Returns:
        sequences  : np.ndarray (n_windows, window, n_features)
        labels     : list of (user_id, date_only) for the last day
                     of each window
    """
    sequences = []
    seq_labels = []

    for user_id, grp in tqdm(
        df.groupby('user_id'),
        desc='Building sequences',
        leave=False
    ):
        grp = grp.sort_values('date_only')
        vals = grp[feature_cols].values.astype(np.float32)
        dates = grp['date_only'].values

        if len(vals) < window:
            # Pad with zeros if user has fewer days than window
            pad = np.zeros(
                (window - len(vals), len(feature_cols)),
                dtype=np.float32
            )
            vals  = np.vstack([pad, vals])
            sequences.append(vals)
            seq_labels.append((user_id, dates[-1]))
        else:
            for i in range(len(vals) - window + 1):
                sequences.append(vals[i:i+window])
                seq_labels.append((user_id, dates[i+window-1]))

    return np.array(sequences, dtype=np.float32), seq_labels


# Use scaled features
fm_scaled = fm.copy()
fm_scaled[FEATURE_COLS] = X_scaled

sequences, seq_labels = build_sequences(
    fm_scaled, FEATURE_COLS, LSTM_WINDOW_DAYS
)

print(f'\n✅ Sequence tensor shape: {sequences.shape}')
print(f'   (n_windows={sequences.shape[0]:,}, '
      f'window={sequences.shape[1]}, '
      f'features={sequences.shape[2]})')
````

````python
# ── LSTM with Attention architecture ─────────────────────────
class ThreatLSTM(nn.Module):
    def __init__(
        self,
        input_dim: int,
        hidden_dim: int = 128,
        num_layers: int = 2,
        dropout: float = 0.2,
    ):
        super().__init__()
        self.lstm = nn.LSTM(
            input_size=input_dim,
            hidden_size=hidden_dim,
            num_layers=num_layers,
            batch_first=True,
            dropout=dropout if num_layers > 1 else 0.0,
        )
        # Attention: score each time step
        self.attention = nn.Linear(hidden_dim, 1)

        # Reconstruction decoder (same dim as input)
        # We use reconstruction loss for unsupervised training
        self.decoder = nn.Sequential(
            nn.Linear(hidden_dim, 64),
            nn.ReLU(),
            nn.Linear(64, input_dim),
        )

    def forward(
        self, x: torch.Tensor
    ) -> tuple:
        """
        Args:
            x: (batch, seq_len, input_dim)
        Returns:
            recon        : (batch, input_dim)  reconstructed last step
            attn_weights : (batch, seq_len)    attention over time
        """
        lstm_out, _ = self.lstm(x)          # (batch, seq, hidden)
        attn_scores  = self.attention(lstm_out)          # (batch, seq, 1)
        attn_weights = torch.softmax(attn_scores, dim=1) # (batch, seq, 1)
        context      = (attn_weights * lstm_out).sum(dim=1)  # (batch, hidden)
        recon        = self.decoder(context)             # (batch, input_dim)
        return recon, attn_weights.squeeze(-1)

    def anomaly_score(
        self, x: torch.Tensor
    ) -> torch.Tensor:
        """MSE between reconstructed context and last time step."""
        with torch.no_grad():
            recon, _ = self.forward(x)
            last_step = x[:, -1, :]          # (batch, input_dim)
            return torch.mean((last_step - recon) ** 2, dim=1)


lstm_model = ThreatLSTM(
    input_dim=INPUT_DIM,
    hidden_dim=LSTM_HIDDEN,
    num_layers=LSTM_LAYERS,
).to(DEVICE)

print('✅ LSTM + Attention architecture:')
print(lstm_model)
lstm_params = sum(p.numel() for p in lstm_model.parameters())
print(f'\n   Total parameters : {lstm_params:,}')
````

````python
# ── Prepare LSTM training data (normal sequences only) ────────
# Map each sequence's label back to its ISO score
iso_lookup = fm.set_index(['user_id', 'date_only'])['iso_score'].to_dict()

seq_iso_scores = np.array([
    iso_lookup.get((uid, date), 0.5)
    for uid, date in seq_labels
])

normal_seq_threshold = np.percentile(seq_iso_scores, 80)
normal_seq_mask = seq_iso_scores <= normal_seq_threshold

X_seq_normal = sequences[normal_seq_mask]
X_seq_tensor = torch.tensor(X_seq_normal, dtype=torch.float32)
X_seq_all    = torch.tensor(sequences,    dtype=torch.float32)

seq_dataset = TensorDataset(X_seq_tensor, X_seq_tensor)
seq_loader  = DataLoader(
    seq_dataset,
    batch_size=LSTM_BATCH_SIZE,
    shuffle=True,
    drop_last=False,
)

print(f'✅ LSTM training sequences prepared.')
print(f'   Total sequences  : {len(sequences):,}')
print(f'   Normal sequences : {X_seq_normal.shape[0]:,} '
      f'({X_seq_normal.shape[0]/len(sequences)*100:.1f}%)')
````

````python
# ── Train LSTM ────────────────────────────────────────────────
lstm_criterion = nn.MSELoss()
lstm_optimizer = optim.Adam(lstm_model.parameters(), lr=LSTM_LR)
lstm_scheduler = optim.lr_scheduler.StepLR(
    lstm_optimizer, step_size=10, gamma=0.5
)

lstm_losses = []

print(f'🧠 Training LSTM for {LSTM_EPOCHS} epochs...\n')
lstm_model.train()

for epoch in range(1, LSTM_EPOCHS + 1):
    epoch_loss = 0.0
    for batch_x, _ in seq_loader:
        batch_x = batch_x.to(DEVICE)
        lstm_optimizer.zero_grad()
        recon, _ = lstm_model(batch_x)
        # Reconstruct the last time step
        loss = lstm_criterion(recon, batch_x[:, -1, :])
        loss.backward()
        nn.utils.clip_grad_norm_(lstm_model.parameters(), 1.0)
        lstm_optimizer.step()
        epoch_loss += loss.item() * len(batch_x)

    avg_loss = epoch_loss / len(X_seq_normal)
    lstm_losses.append(avg_loss)
    lstm_scheduler.step()

    if epoch % 5 == 0 or epoch == 1:
        print(f'   Epoch {epoch:>3}/{LSTM_EPOCHS}  '
              f'Loss: {avg_loss:.6f}  '
              f'LR: {lstm_scheduler.get_last_lr()[0]:.6f}')

print(f'\n✅ LSTM training complete.')
print(f'   Final loss: {lstm_losses[-1]:.6f}')
````

````python
# ── Compute LSTM anomaly scores for all sequences ─────────────
lstm_model.eval()

lstm_errors = []
with torch.no_grad():
    for i in range(0, len(X_seq_all), LSTM_BATCH_SIZE):
        batch = X_seq_all[i:i+LSTM_BATCH_SIZE].to(DEVICE)
        err   = lstm_model.anomaly_score(batch)
        lstm_errors.append(err.cpu().numpy())

lstm_errors_raw = np.concatenate(lstm_errors)

mm_scaler_lstm = MinMaxScaler()
lstm_scores_norm = mm_scaler_lstm.fit_transform(
    lstm_errors_raw.reshape(-1, 1)
).flatten()

# Map sequence scores back to the feature matrix
# Each sequence maps to (user_id, date_only) of its last day
seq_score_df = pd.DataFrame({
    'user_id'    : [lbl[0] for lbl in seq_labels],
    'date_only'  : [lbl[1] for lbl in seq_labels],
    'lstm_score' : lstm_scores_norm,
})
# Keep max score per user-day (in case of overlapping windows)
seq_score_df = (
    seq_score_df
    .groupby(['user_id', 'date_only'])['lstm_score']
    .max()
    .reset_index()
)

fm = fm.merge(seq_score_df, on=['user_id', 'date_only'], how='left')
fm['lstm_score'] = fm['lstm_score'].fillna(0)

print(f'✅ LSTM scores mapped back to feature matrix.')
print(f'   Score range : [{lstm_scores_norm.min():.4f}, '
      f'{lstm_scores_norm.max():.4f}]')
print(f'   Flagged rows (score > {THRESHOLD_ELEVATED}): '
      f'{(fm["lstm_score"] > THRESHOLD_ELEVATED).sum():,}')
````

````python
# ── Visualise LSTM training loss + score distribution ─────────
fig, axes = plt.subplots(1, 2, figsize=(14, 4))

# Training loss
axes[0].plot(range(1, LSTM_EPOCHS + 1), lstm_losses,
             color='#DA70D6', linewidth=2)
axes[0].fill_between(range(1, LSTM_EPOCHS + 1), lstm_losses,
                     alpha=0.2, color='#DA70D6')
axes[0].set_title('LSTM Training Loss (MSE)',
                  fontsize=12, fontweight='bold')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('MSE Loss')
axes[0].grid(alpha=0.3)

# Score distribution
axes[1].hist(fm['lstm_score'], bins=80,
             color='#DA70D6', alpha=0.75, edgecolor='none')
for thresh, color, label in [
    (THRESHOLD_WATCH,    '#FFD700', 'Watch'),
    (THRESHOLD_ELEVATED, '#FF6B35', 'Elevated'),
    (THRESHOLD_CRITICAL, '#FF4444', 'Critical'),
]:
    axes[1].axvline(thresh, color=color,
                   linestyle='--', linewidth=1.5,
                   label=label)
axes[1].set_title('LSTM Score Distribution',
                  fontsize=12, fontweight='bold')
axes[1].set_xlabel('LSTM Anomaly Score')
axes[1].set_ylabel('Count')
axes[1].legend(fontsize=9)
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.savefig('lstm_results.png', dpi=120, bbox_inches='tight')
plt.show()
print('✅ LSTM results saved → lstm_results.png')
````

## 2.4 Risk Score Fusion & Alert Thresholding

We combine the three model scores into a single **RiskScore** per user per day
using a weighted ensemble:

$$\text{RiskScore} = 0.35 \times \text{iso\_score} + 0.35 \times \text{ae\_score} + 0.30 \times \text{lstm\_score}$$

All inputs are already normalised to [0, 1]. The fused score is then classified
into four alert levels:

| Score Range | Level | Action |
|-------------|-------|--------|
| 0.00 – 0.40 | 🟢 Normal | No action |
| 0.40 – 0.65 | 🟡 Watch | Enhanced logging |
| 0.65 – 0.80 | 🟠 Elevated | LLM analysis triggered |
| 0.80 – 1.00 | 🔴 Critical | Immediate SOC alert |

````python
# ── Compute fused risk score ──────────────────────────────────
fm['risk_score'] = (
    WEIGHT_ISO  * fm['iso_score']  +
    WEIGHT_AE   * fm['ae_score']   +
    WEIGHT_LSTM * fm['lstm_score']
)

# Classify into alert levels
def classify_alert(score: float) -> str:
    if score >= THRESHOLD_CRITICAL:
        return 'critical'
    elif score >= THRESHOLD_ELEVATED:
        return 'elevated'
    elif score >= THRESHOLD_WATCH:
        return 'watch'
    return 'normal'

fm['alert_level'] = fm['risk_score'].apply(classify_alert)

level_counts = fm['alert_level'].value_counts()
print('✅ Risk score fusion complete.')
print(f'\n   Risk score range : [{fm["risk_score"].min():.4f}, '
      f'{fm["risk_score"].max():.4f}]')
print(f'   Risk score mean  : {fm["risk_score"].mean():.4f}')
print('\n   Alert level distribution:')
for level in ['critical', 'elevated', 'watch', 'normal']:
    count = level_counts.get(level, 0)
    pct   = count / len(fm) * 100
    bar   = '█' * int(pct / 2)
    print(f'   {level:<10}: {count:>7,} rows  ({pct:>5.1f}%)  {bar}')
````

````python
# ── Per-user aggregated risk scores ───────────────────────────
user_risk = (
    fm.groupby('user_id')
    .agg(
        max_risk_score   = ('risk_score',  'max'),
        mean_risk_score  = ('risk_score',  'mean'),
        max_iso_score    = ('iso_score',   'max'),
        max_ae_score     = ('ae_score',    'max'),
        max_lstm_score   = ('lstm_score',  'max'),
        critical_days    = ('alert_level', lambda x: (x == 'critical').sum()),
        elevated_days    = ('alert_level', lambda x: (x == 'elevated').sum()),
        watch_days       = ('alert_level', lambda x: (x == 'watch').sum()),
        total_days       = ('risk_score',  'count'),
    )
    .reset_index()
    .sort_values('max_risk_score', ascending=False)
)

# Add role/department
ldap_cols = ['user_id'] + [
    c for c in ['role', 'department'] if c in fm.columns
]
user_risk = user_risk.merge(
    fm[ldap_cols].drop_duplicates('user_id'),
    on='user_id', how='left'
)

print(f'✅ Per-user risk aggregation complete.')
print(f'   Total users scored: {len(user_risk):,}')
print(f'\n   Top 10 highest-risk users:')
user_risk.head(10)[[
    'user_id', 'max_risk_score', 'critical_days',
    'elevated_days', 'role'
]]
````

## 2.5 SHAP Explainability

SHAP (SHapley Additive exPlanations) decomposes each model prediction into
per-feature contributions. We apply it to the **Isolation Forest** (tree-based,
natively supported by SHAP's `TreeExplainer`) to explain why each flagged
user received a high anomaly score.

Every SOC alert will include the **top-5 contributing features** ranked by
absolute SHAP value — bridging the gap between black-box ML and analyst-
interpretable evidence.

````python
# ── Compute SHAP values for Isolation Forest ──────────────────
print('🔍 Computing SHAP values for Isolation Forest...')
print('   (This may take 1–3 minutes depending on dataset size)\n')

# Use a background sample for efficiency
# SHAP TreeExplainer works best with a representative background
N_BACKGROUND = min(500, len(X_scaled))
bg_idx = np.random.choice(len(X_scaled), N_BACKGROUND, replace=False)
X_background = X_scaled[bg_idx]

explainer = shap.TreeExplainer(
    iso_forest,
    data=X_background,
    feature_perturbation='interventional',
)

# Compute SHAP values for flagged rows only (elevated + critical)
flagged_mask = fm['risk_score'] >= THRESHOLD_ELEVATED
X_flagged    = X_scaled[flagged_mask]

print(f'   Computing SHAP for {flagged_mask.sum():,} flagged rows '
      f'(risk >= {THRESHOLD_ELEVATED})...')

shap_values = explainer.shap_values(X_flagged)

# shap_values shape: (n_flagged, n_features)
print(f'\n✅ SHAP values computed: {shap_values.shape}')
````

````python
# ── Global SHAP summary: mean absolute impact per feature ─────
shap_abs_mean = np.abs(shap_values).mean(axis=0)
shap_importance = pd.DataFrame({
    'feature'    : FEATURE_COLS,
    'shap_impact': shap_abs_mean,
}).sort_values('shap_impact', ascending=False).reset_index(drop=True)

print('📊 Top 20 features by mean absolute SHAP impact:')
print(shap_importance.head(20).to_string(index=False))
````

````python
# ── Plot global SHAP feature importance ──────────────────────
top_n = 20
top_shap = shap_importance.head(top_n)

fig = px.bar(
    top_shap[::-1],
    x='shap_impact',
    y='feature',
    orientation='h',
    title=f'Top {top_n} Features — Mean Absolute SHAP Impact '
          f'(Isolation Forest, Flagged Users)',
    labels={
        'shap_impact': 'Mean |SHAP Value|',
        'feature': 'Feature'
    },
    color='shap_impact',
    color_continuous_scale='Reds',
    height=550,
)
fig.update_layout(
    plot_bgcolor='#0D1B2A',
    paper_bgcolor='#0D1B2A',
    font_color='white',
    title_font_size=13,
    coloraxis_showscale=False,
)
fig.show()
print('✅ Global SHAP importance plot rendered.')
````

````python
# ── Per-user SHAP explanation helper ─────────────────────────
flagged_fm = fm[flagged_mask].reset_index(drop=True)

def explain_user(
    user_id: str,
    top_n: int = 5
) -> pd.DataFrame:
    """
    Return the top-N SHAP feature contributions for the
    highest-risk day of a given user.

    Args:
        user_id : The user to explain.
        top_n   : Number of top features to return.

    Returns:
        DataFrame with feature, shap_value, feature_value,
        and direction columns.
    """
    user_rows = flagged_fm[flagged_fm['user_id'] == user_id]
    if user_rows.empty:
        print(f'⚠️  User {user_id} not found in flagged set.')
        return pd.DataFrame()

    # Pick the row with the highest risk score
    worst_idx  = user_rows['risk_score'].idxmax()
    row_pos    = flagged_fm.index.get_loc(worst_idx)
    shap_row   = shap_values[row_pos]
    feat_row   = X_flagged[row_pos]

    explanation = pd.DataFrame({
        'feature'       : FEATURE_COLS,
        'shap_value'    : shap_row,
        'feature_value' : feat_row,
    })
    explanation['abs_shap']  = explanation['shap_value'].abs()
    explanation['direction'] = explanation['shap_value'].apply(
        lambda v: '🔴 raises risk' if v > 0 else '🟢 lowers risk'
    )
    explanation = (
        explanation
        .sort_values('abs_shap', ascending=False)
        .head(top_n)
        .reset_index(drop=True)
    )

    worst_date  = flagged_fm.loc[worst_idx, 'date_only']
    worst_score = flagged_fm.loc[worst_idx, 'risk_score']
    print(f'\n🔍 SHAP Explanation for user: {user_id}')
    print(f'   Worst day   : {worst_date.date()}')
    print(f'   Risk score  : {worst_score:.4f}')
    print(f'   Alert level : '
          f'{classify_alert(worst_score).upper()}')
    print()
    return explanation[[
        'feature', 'shap_value',
        'feature_value', 'direction'
    ]]


# Demo: explain the top-risk user
top_user = user_risk.iloc[0]['user_id']
explain_user(top_user, top_n=5)
````

````python
# ── Waterfall-style SHAP bar chart for top user ───────────────
top_user_explanation = explain_user(top_user, top_n=10)

if not top_user_explanation.empty:
    colors = [
        '#FF4444' if v > 0 else '#00C853'
        for v in top_user_explanation['shap_value']
    ]
    fig = go.Figure(go.Bar(
        x=top_user_explanation['shap_value'],
        y=top_user_explanation['feature'],
        orientation='h',
        marker_color=colors,
        text=top_user_explanation['shap_value'].round(4),
        textposition='outside',
    ))
    fig.update_layout(
        title=f'SHAP Waterfall — User: {top_user} '
              f'(Risk Score: '
              f'{user_risk.iloc[0]["max_risk_score"]:.4f})',
        xaxis_title='SHAP Value (impact on anomaly score)',
        yaxis_title='Feature',
        height=450,
        plot_bgcolor='#0D1B2A',
        paper_bgcolor='#0D1B2A',
        font_color='white',
        title_font_size=13,
        yaxis={'categoryorder': 'total ascending'},
    )
    fig.add_vline(
        x=0, line_color='white',
        line_width=1, line_dash='solid'
    )
    fig.show()
    print('✅ SHAP waterfall chart rendered.')
    print('\n💡 Red bars = features that INCREASE anomaly score.')
    print('   Green bars = features that DECREASE anomaly score.')
````

````python
# ── Attach top-5 SHAP features to each flagged user-day ───────
print('📎 Attaching top-5 SHAP features to flagged user-day records...')

top5_records = []
for i in range(len(flagged_fm)):
    shap_row  = shap_values[i]
    top5_idx  = np.argsort(np.abs(shap_row))[::-1][:5]
    top5_feats = [
        {
            'feature'    : FEATURE_COLS[j],
            'shap_value' : round(float(shap_row[j]), 5),
            'feat_value' : round(float(X_flagged[i][j]), 5),
        }
        for j in top5_idx
    ]
    top5_records.append(str(top5_feats))

flagged_fm = flagged_fm.copy()
flagged_fm['shap_top5'] = top5_records

print(f'✅ SHAP top-5 attached to {len(flagged_fm):,} flagged rows.')
flagged_fm[[
    'user_id', 'date_only', 'risk_score',
    'alert_level', 'shap_top5'
]].head(3)
````
